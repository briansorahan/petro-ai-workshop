Training and Testing Data

To evaluate our models we split our dataset into a training set and a test set. If you remember from the supervised learning example for apples and oranges we wanted to test how well our machine had learned to tell the difference by testing the machine with apples and oranges it hadn't previously seen.

For our example we are going to train a machine to predict a compressional wave sonic log using some drilling data and a gamma ray log. We will split VertB into a training and testing set and evaluate the performance of the model within the VertB dataset. We will then use this model to predict the compressional wave sonic log in VertC and evaluate our performance. The goal is a fair evaluation of the system.



```python
VertA=VertA[['DEPT','DT35','GMSG','ROP_-_FAST:1','BIT_WEIGHT','ROTARY_RPM']]
VertB=VertB[['DEPT','DT35','GMGC','ROP_-_FAST:1','BIT_WEIGHT','ROTARY_RPM']]
VertC=VertC[['DEPT','DTCO','GR','ROP_-_FAST:1','BIT_WEIGHT','ROTARY_RPM']]
VertD=VertD[['DEPT','MCDT','GMGC','ROP_-_FAST:1','BIT_WEIGHT','ROTARY_RPM']]
VertE=VertE[['DEPT','DT','GRTO','ROP','WOB','RPM']]
VertA = VertA.rename(index=str,columns={"DEPT":"DEPT","DT35":"DT","GMSG":"GR","ROP_-_FAST:1":"ROP","BIT_WEIGHT":"WOB","ROTARY_RPM":"RPM"})
VertB = VertB.rename(index=str,columns={"DEPT":"DEPT","DT35":"DT","GMGC":"GR","ROP_-_FAST:1":"ROP","BIT_WEIGHT":"WOB","ROTARY_RPM":"RPM"})
VertC = VertC.rename(index=str,columns={"DEPT":"DEPT","DTCO":"DT","GR":"GR","ROP_-_FAST:1":"ROP","BIT_WEIGHT":"WOB","ROTARY_RPM":"RPM"})
VertD = VertD.rename(index=str,columns={"DEPT":"DEPT","MCDT":"DT","GMGC":"GR","ROP_-_FAST:1":"ROP","BIT_WEIGHT":"WOB","ROTARY_RPM":"RPM"})
VertE = VertE.rename(index=str,columns={"DEPT":"DEPT","DT":"DT","GRTO":"GR","ROP":"ROP","WOB":"WOB","RPM":"RPM"})


frames = [VertA,VertB,VertC,VertD]
mergeABCD = pd.concat(frames,ignore_index=True)
```

```python
# Let's begin by setting up our subset datasets.

X_mergeABCD = mergeABCD[['ROP','WOB','RPM','GR']]
y_mergeABCD = mergeABCD['DT']

X_VertE = VertE[['ROP','WOB','RPM','GR']]
y_VertE = VertE['DT']
```

```python
# We will now divide out data into a training and test sets using a rather handy function.
# http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

from sklearn.model_selection import train_test_split

train_X_mergeABCD, test_X_mergeABCD, train_y_mergeABCD, test_y_mergeABCD = train_test_split(X_mergeABCD, y_mergeABCD, 
                                                    train_size=0.9,
                                                    test_size=0.1,
                                                    random_state=42)
```

```python
# We will use a neighbour based method for regression.

from sklearn.neighbors import KNeighborsRegressor

kneighbor_regression = KNeighborsRegressor(n_neighbors=5,n_jobs=-1)

kneighbor_regression.fit(train_X_mergeABCD, train_y_mergeABCD)
mergeABCD_test = kneighbor_regression.predict(test_X_mergeABCD)
```

```python
VertE_Pred_DTCO_KNR = kneighbor_regression.predict(X_VertE)
```

```python
# Plot results

x = VertE['DEPT']
y = VertE['DT']
z = VertE_Pred_DTCO_KNR

p = figure(title='VertE Predicted DTCO using KNR', plot_width=600, plot_height=300)
r = p.line(x, y, color='green', line_width=0.5,legend='DT')
r1 = p.line(x, z, color='blue', line_width=0.5,legend='Pred. DT')
p.xaxis.axis_label = 'Depth (ft)'
p.yaxis.axis_label = 'Delta Time (ms/ft)'
p.legend.location = 'top_left'
show(p)
```

```python
# Can we make this prediction better? Let's try a different method in sklearn.
# http://scikit-learn.org/stable/modules/generated/sklearn.svm.SVR.html

from sklearn.svm import SVR
svr = SVR(kernel='linear',C=1232)
```

```python
# SVR works much faster when data is normalized.

from sklearn.preprocessing import Normalizer

scaler = Normalizer().fit(mergeABCD[['ROP','WOB','RPM','GR']])
normalized_mergeABCD = scaler.transform(mergeABCD[['ROP','WOB','RPM','GR']])

scaler = Normalizer().fit(VertE[['ROP','WOB','RPM','GR']])
normalized_VertE = scaler.transform(VertE[['ROP','WOB','RPM','GR']])

# Convert back to pandas dataframes
normalized_mergeABCD = pd.DataFrame(normalized_mergeABCD,columns=[['ROP','WOB','RPM','GR']])
normalized_VertE = pd.DataFrame(normalized_VertE,columns=[['ROP','WOB','RPM','GR']])
```

```python
# Split the normalized data into training and test sets

train_X_mergeABCD, test_X_mergeABCD, train_y_mergeABCD, test_y_mergeABCD = train_test_split(normalized_mergeABCD, y_mergeABCD, 
                                                    train_size=0.9,
                                                    test_size=0.1,
                                                    random_state=42)

train_X_VertE, test_X_VertE, train_y_VertE, test_y_VertE = train_test_split(normalized_VertE, y_VertE, 
                                                    train_size=0.9,
                                                    test_size=0.1,
                                                    random_state=42)
```

We are now going to start to tune our model. There isn't really a general rule for finding the sweet spot. We will start by ranking our features based on what the model determines is important. We will then perform a grid step search looking for the right settings to get the 'best' model performance.

```python
# SVR works much faster when data is normalized.

scaler = Normalizer().fit(VertB[['ROP_-_FAST:1', 'BIT_WEIGHT','ROTARY_RPM', 'GMGC','NPRL:1','DEN','ANNULAR_VELOCITY','WASHOUT_FACTOR']])
normalized_VertB = scaler.transform(VertB[['ROP_-_FAST:1', 'BIT_WEIGHT','ROTARY_RPM', 'GMGC','NPRL:1','DEN','ANNULAR_VELOCITY','WASHOUT_FACTOR']])

scaler = Normalizer().fit(VertC[['ROP_-_FAST:1', 'BIT_WEIGHT','ROTARY_RPM', 'GR','NPOR','RHOB','ANNULAR_VELOCITY','WASHOUT_FACTOR']])
normalized_VertC = scaler.transform(VertC[['ROP_-_FAST:1', 'BIT_WEIGHT','ROTARY_RPM', 'GR','NPOR','RHOB','ANNULAR_VELOCITY','WASHOUT_FACTOR']])

# Convert back to pandas dataframes
normalized_VertB = pd.DataFrame(normalized_VertB,columns=[['ROP_-_FAST:1', 'BIT_WEIGHT','ROTARY_RPM', 'GMGC','NPRL:1','DEN','ANNULAR_VELOCITY','WASHOUT_FACTOR']])
normalized_VertC = pd.DataFrame(normalized_VertC,columns=[['ROP_-_FAST:1', 'BIT_WEIGHT','ROTARY_RPM', 'GR','NPOR','RHOB','ANNULAR_VELOCITY','WASHOUT_FACTOR']])
```

The recursive feature elimination (RFE) module selects subsets of features. The estimator is trained on all the features in question and an importance ranking is obtained. The least important feature is dropped recursively until the specified number of features is obtained.

```python
# http://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.RFE.html
from sklearn.feature_selection import RFE

select = RFE(svr,n_features_to_select=4)
```

```python
select.fit(normalized_VertB, y_VertB)
normalized_VertB_FS = select.transform(normalized_VertB)
print(normalized_VertB.shape)
print(normalized_VertB_FS.shape)
```

```python
normalized_VertB_FS = pd.DataFrame(normalized_VertB_FS,columns=[['BIT_WEIGHT','GMGC', 'NPRL:1','DEN']])
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score, KFold

cv = KFold(n_splits=5, random_state=42, shuffle=True)

param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000, 1232], 'gamma': [0.0001, 0.001, 0.01, 0.1, 1, 10]}

grid = GridSearchCV(SVR(), param_grid=param_grid, cv=cv, verbose=1, return_train_score=True, n_jobs=-1)
```
