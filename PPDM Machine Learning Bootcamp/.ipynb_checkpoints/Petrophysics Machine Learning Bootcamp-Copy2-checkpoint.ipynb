Training and Testing Data

To evaluate our models we split our dataset into a training set and a test set. If you remember from the supervised learning example for apples and oranges we wanted to test how well our machine had learned to tell the difference by testing the machine with apples and oranges it hadn't previously seen.

For our example we are going to train a machine to predict a compressional wave sonic log using some drilling data and a gamma ray log. We will split VertB into a training and testing set and evaluate the performance of the model within the VertB dataset. We will then use this model to predict the compressional wave sonic log in VertC and evaluate our performance. The goal is a fair evaluation of the system.



```python
# Let's begin by setting up our VertB and VertC subset datasets.

X_VertB = VertB[['ROP_-_FAST:1', 'BIT_WEIGHT','ROTARY_RPM', 'GMGC']]
y_VertB = VertB['DT35']

X_VertC = VertC[['ROP_-_FAST:1', 'BIT_WEIGHT','ROTARY_RPM', 'GR']]
y_VertC = VertC['DTCO']
```

```python
# First we will use VertB to predict VertC and then VertC to predict VertB.
# We will now divide out data into a training and test sets using a rather handy function.
# http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

from sklearn.model_selection import train_test_split

train_X_VertB, test_X_VertB, train_y_VertB, test_y_VertB = train_test_split(X_VertB, y_VertB, 
                                                    train_size=0.9,
                                                    test_size=0.1,
                                                    random_state=42)

train_X_VertC, test_X_VertC, train_y_VertC, test_y_VertC = train_test_split(X_VertC, y_VertC, 
                                                    train_size=0.9,
                                                    test_size=0.1,
                                                    random_state=42)
```

```python
# We will use a neighbour based method for regression.

from sklearn.neighbors import KNeighborsRegressor

kneighbor_regression = KNeighborsRegressor(n_neighbors=5,n_jobs=-1)

kneighbor_regression.fit(train_X_VertB, train_y_VertB)
VertB_test = kneighbor_regression.predict(test_X_VertB)
```

```python
VertC_Pred_DTCO_KNR = kneighbor_regression.predict(X_VertC)
```

```python
# Plot results

x = VertC['DEPT']
y = VertC['DTCO']
z = VertC_Pred_DTCO_KNR

p = figure(title='VertC Predicted DTCO using KNR', plot_width=600, plot_height=300)
r = p.line(x, y, color='green', line_width=0.5,legend='DTCO')
r1 = p.line(x, z, color='blue', line_width=0.5,legend='Pred. DTCO')
p.xaxis.axis_label = 'Depth (ft)'
p.yaxis.axis_label = 'Delta Time (ms/ft)'
p.legend.location = 'top_left'
show(p)
```

```python
# Now train on VertC to predict VertB
kneighbor_regression.fit(train_X_VertC, train_y_VertC)

VertC_test = kneighbor_regression.predict(test_X_VertC)
```

```python
VertB_Pred_DT35_KNR = kneighbor_regression.predict(X_VertB)
```

```python
x = VertB['DEPT']
y = VertB['DT35']
z = VertB_Pred_DT35_KNR

p = figure(title='VertB Predicted DT35 using KNR', plot_width=600, plot_height=300)
r = p.line(x, y, color='green', line_width=0.25,legend='DT35')
r1 = p.line(x, z, color='blue', line_width=0.25,legend='Pred. DT35')
p.xaxis.axis_label = 'Depth (ft)'
p.yaxis.axis_label = 'Delta Time (ms/ft)'
p.legend.location = 'top_left'
show(p)
```

```python
# We will now use mean absolute error to compare performance.
from sklearn.metrics import mean_absolute_error

error_VertB_model = mean_absolute_error(y_VertC, VertC_Pred_DTCO_KNR)
error_VertC_model = mean_absolute_error(y_VertB, VertB_Pred_DT35_KNR)
```

```python
# Can we make this prediction better? Let's try a different method in sklearn.
# http://scikit-learn.org/stable/modules/generated/sklearn.svm.SVR.html

from sklearn.svm import SVR
svr = SVR(kernel='linear',C=1232)
```

```python
# SVR works much faster when data is normalized.

from sklearn.preprocessing import Normalizer

scaler = Normalizer().fit(VertB[['ROP_-_FAST:1','BIT_WEIGHT','ROTARY_RPM','GMGC']])
normalized_VertB = scaler.transform(VertB[['ROP_-_FAST:1','BIT_WEIGHT','ROTARY_RPM','GMGC']])

scaler = Normalizer().fit(VertC[['ROP_-_FAST:1','BIT_WEIGHT','ROTARY_RPM','GR']])
normalized_VertC = scaler.transform(VertC[['ROP_-_FAST:1','BIT_WEIGHT','ROTARY_RPM','GR']])

# Convert back to pandas dataframes
normalized_VertB = pd.DataFrame(normalized_VertB,columns=[['ROP_-_FAST:1','BIT_WEIGHT','ROTARY_RPM','GMGC']])
normalized_VertC = pd.DataFrame(normalized_VertC,columns=[['ROP_-_FAST:1','BIT_WEIGHT','ROTARY_RPM','GR']])
```

```python
# Split the normalized data into training and test sets

train_X_VertB, test_X_VertB, train_y_VertB, test_y_VertB = train_test_split(normalized_VertB, y_VertB, 
                                                    train_size=0.9,
                                                    test_size=0.1,
                                                    random_state=42)

train_X_VertC, test_X_VertC, train_y_VertC, test_y_VertC = train_test_split(normalized_VertC, y_VertC, 
                                                    train_size=0.9,
                                                    test_size=0.1,
                                                    random_state=42)
```

```python
# Fit VertB and predict VertC
VertBtoVertC_DTCO = svr.fit(train_X_VertB, train_y_VertB)
VertC_Pred_DTCO = VertBtoVertC_DTCO.predict(normalized_VertC)
```

```python
x = VertC['DEPT']
y = VertC['DTCO']
z = VertC_Pred_DTCO
z1 = VertC_Pred_DTCO_KNR

p = figure(title='VertC Predicted DTCO using SVR', plot_width=600, plot_height=300)
r = p.line(x, y, color='green', line_width=0.5,legend='DTCO')
r1 = p.line(x, z, color='blue', line_width=0.5,legend='Pred. DTCO SVR')
r2 = p.line(x, z1, color='red', line_width=0.5,legend='Pred. DTCO KNR')

p.xaxis.axis_label = 'Depth (ft)'
p.yaxis.axis_label = 'Delta Time (ms/ft)'
p.legend.location = 'top_left'
show(p)
```

```python
error_VertB_model = mean_absolute_error(y_VertC, VertC_Pred_DTCO)
error_VertB_model
```

```python
# Fit VertC and predict VertB
VertCtoVertB_DT35 = svr.fit(train_X_VertC, train_y_VertC)
VertB_Pred_DT35 = VertCtoVertB_DT35.predict(normalized_VertB)
```

```python
x = VertB['DEPT']
y = VertB['DT35']
z = VertB_Pred_DT35
z1 = VertB_Pred_DT35_KNR

p = figure(title='VertB Predicted DT35 using SVR', plot_width=600, plot_height=300)
r = p.line(x, y, color='green', line_width=0.5,legend='DT35')
r1 = p.line(x, z, color='blue', line_width=0.5,legend='Pred. DT35 SVR')
r2 = p.line(x, z1, color='red', line_width=0.5,legend='Pred. DT35 KNR')

p.xaxis.axis_label = 'Depth (ft)'
p.yaxis.axis_label = 'Delta Time (ms/ft)'
p.legend.location = 'top_left'
show(p)
```

We are now going to start to tune our model. There isn't really a general rule for finding the sweet spot. We will start by ranking our features based on what the model determines is important. We will then perform a grid step search looking for the right settings to get the 'best' model performance. 

For recursive feature elimination we need to add back all the overlapping data between VertB and VertC.

```python
X_VertB = VertB[['ROP_-_FAST:1', 'BIT_WEIGHT','ROTARY_RPM', 'GMGC','NPRL:1','DEN','ANNULAR_VELOCITY','WASHOUT_FACTOR']]
y_VertB = VertB['DT35']

X_VertC = VertC[['ROP_-_FAST:1', 'BIT_WEIGHT','ROTARY_RPM', 'GR','NPOR','RHOB','ANNULAR_VELOCITY','WASHOUT_FACTOR']]
y_VertC = VertC['DTCO']
```

```python
# SVR works much faster when data is normalized.

scaler = Normalizer().fit(VertB[['ROP_-_FAST:1', 'BIT_WEIGHT','ROTARY_RPM', 'GMGC','NPRL:1','DEN','ANNULAR_VELOCITY','WASHOUT_FACTOR']])
normalized_VertB = scaler.transform(VertB[['ROP_-_FAST:1', 'BIT_WEIGHT','ROTARY_RPM', 'GMGC','NPRL:1','DEN','ANNULAR_VELOCITY','WASHOUT_FACTOR']])

scaler = Normalizer().fit(VertC[['ROP_-_FAST:1', 'BIT_WEIGHT','ROTARY_RPM', 'GR','NPOR','RHOB','ANNULAR_VELOCITY','WASHOUT_FACTOR']])
normalized_VertC = scaler.transform(VertC[['ROP_-_FAST:1', 'BIT_WEIGHT','ROTARY_RPM', 'GR','NPOR','RHOB','ANNULAR_VELOCITY','WASHOUT_FACTOR']])

# Convert back to pandas dataframes
normalized_VertB = pd.DataFrame(normalized_VertB,columns=[['ROP_-_FAST:1', 'BIT_WEIGHT','ROTARY_RPM', 'GMGC','NPRL:1','DEN','ANNULAR_VELOCITY','WASHOUT_FACTOR']])
normalized_VertC = pd.DataFrame(normalized_VertC,columns=[['ROP_-_FAST:1', 'BIT_WEIGHT','ROTARY_RPM', 'GR','NPOR','RHOB','ANNULAR_VELOCITY','WASHOUT_FACTOR']])
```

The recursive feature elimination (RFE) module selects subsets of features. The estimator is trained on all the features in question and an importance ranking is obtained. The least important feature is dropped recursively until the specified number of features is obtained.

```python
# http://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.RFE.html
from sklearn.feature_selection import RFE

select = RFE(svr,n_features_to_select=4)
```

```python
select.fit(normalized_VertB, y_VertB)
normalized_VertB_FS = select.transform(normalized_VertB)
print(normalized_VertB.shape)
print(normalized_VertB_FS.shape)
```

```python
normalized_VertB_FS = pd.DataFrame(normalized_VertB_FS,columns=[['BIT_WEIGHT','GMGC', 'NPRL:1','DEN']])
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score, KFold

cv = KFold(n_splits=5, random_state=42, shuffle=True)

param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000, 1232], 'gamma': [0.0001, 0.001, 0.01, 0.1, 1, 10]}

grid = GridSearchCV(SVR(), param_grid=param_grid, cv=cv, verbose=1, return_train_score=True, n_jobs=-1)
```

```python
# Take the hyperparameter tuned model with RFE to predict DTCO for VertC.
hypertune_SVR = grid.predict(normalized_VertC[['BIT_WEIGHT','GR', 'NPOR','RHOB']])
```

```python
x = VertC['DEPT']
y = VertC['DTCO']
z = VertC_Pred_DTCO
z1 = VertC_Pred_DTCO_KNR
z2 = hypertune_SVR

p = figure(title='VertC Predicted DTCO using SVR', plot_width=600, plot_height=300)
r = p.line(x, y, color='green', line_width=0.5,legend='DTCO')
r1 = p.line(x, z, color='blue', line_width=0.5,legend='Pred. DTCO SVR')
r2 = p.line(x, z1, color='red', line_width=0.5,legend='Pred. DTCO KNR')
r3 = p.line(x, z2, color='orange', line_width=0.5,legend='hypertune_SVR')

p.xaxis.axis_label = 'Depth (ft)'
p.yaxis.axis_label = 'Delta Time (ms/ft)'
p.legend.location = 'top_left'
show(p)
```

```python
# Calculate MAE so we can compare model performance before and after tuning.
error_VertB_hypertune_model = mean_absolute_error(y_VertC, hypertune_SVR)
print(error_VertB_hypertune_model)
print(error_VertB_model)
```
